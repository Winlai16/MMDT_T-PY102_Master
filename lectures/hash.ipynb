```python
def HashInsert(table: list[list[tuple]], key, value) -> None:
    index = hash(key) % len(table)
    bucket = table[index]

    for i, (k, v) in enumerate(bucket):
        if k == key:
            bucket[i] = (key, value)
            return

    bucket.append((key, value))

def h(key):
    return key % 3

table = [[], [],[]]
HashInsert(table, 4, "A")
HashInsert(table, 1, "B")

HashInsert(table, 3, "A")
HashInsert(table, 6, "B")
HashInsert(table, 9, "C")
HashInsert(table, 12, "D")

table[1][0]
```

```python
def HashGet(table: list[list[tuple]], key):
    index = hash(key) % len(table)
    bucket = table[index]

    for k, v in bucket:
        if k == key:
            return v

    return None

HashGet(table, 9)
```

```python
def HashRemove(table: list[list[tuple]], key) -> bool:
    index = hash(key) % len(table)
    bucket = table[index]

    for i, (k, v) in enumerate(bucket):
        if k == key:
            bucket.pop(i)
            return True

    return False
```
